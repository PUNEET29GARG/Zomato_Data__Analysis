````python
# Price range ,Has Table booking,Has Online delivery,,"Average Cost for two","Restaurant Name"--->satoring into new data set
filtered_data=data[["Price range","Has Table booking","Has Online delivery","Average Cost for two","Restaurant Name"]]
filtered_data
````

````python
#where "Has Table booking"=="Yes"
Table_booking_yes=filtered_data[filtered_data["Has Table booking"]=="Yes"]
````

````python
#finding mean price range for individual resturants where "Has Table booking"=="Yes"
Price_rangeTable_booking_yes=Table_booking_yes.groupby("Restaurant Name")["Price range"].mean()
#Price_rangeTable_booking_yes
````

````python
x=list(Price_rangeTable_booking_yes)
mean_priceRange_for_reaturants_havingTableBooking=sum(x)/len(x)
mean_priceRange_for_reaturants_havingTableBooking
````

````python
#creating a new data set where Has Online delivery"=="Yes"
have_online_booking =filtered_data[filtered_data["Has Online delivery"]=="Yes"]
````

````python
#finding mean price range for individual resturants where Has Online delivery"=="Yes"
Price_range_have_online_booking=have_online_booking.groupby("Restaurant Name")["Price range"].mean()
#Price_range_have_online_booking
````

````python
_x=list(Price_range_have_online_booking)
mean_priceRange_for_reaturants_HasOnline_delivery=sum(_x)/len(_x)
mean_priceRange_for_reaturants_HasOnline_delivery
````

**Let's do the comparision b/w both cuisines i.e "Rest of India", "Delhi NCR"**<br>
**For that let's concatenate both data into one , just taking top 10 entries from both**

````python
#concatenating alon axis=1 i.e column
top_10_cuisines = pd.concat([Cuisines_of_Rest_of_India[:10], Cuisines_of_Delhi_NCR[:10]], axis=1)
#renaming column names
top_10_cuisines.columns = ["Rest of India", "Delhi NCR"]
top_10_cuisines
````

**Let's plot bar graph for common cuisines only from top 10 entries to get an idea that which cuisine is more famous where i.e either in rest of india or delhi NCR**

````python
#dropping rows with null values in order to get common cuisines
common_cuisines=top_10_cuisines.dropna()
common_cuisines
````

````python
#plotting bar graph
common_cuisines.plot(kind="bar")
#giving title to the graph
plt.title("Top 10 Cuisines Comparison")
````
